```python
from skimage.io import imread, imsave
from skimage.transform import resize
from skimage.filters import median

import numpy as np
import matplotlib.pyplot as plt
```

```python
img = imread('./app.png')

plt.imshow(img)
plt.show()
```

```python
mask = ~(img == 255).all(2)
mask = median(mask, footprint=np.ones((9,9)))
mask = (mask).astype('uint8') * 255
plt.imshow(mask)
plt.show()
mask = mask[..., np.newaxis]
```

```python
img = np.concat([img, mask], axis=-1)
img = np.uint8(resize(img, (256,256), order=1) * 255)
plt.imshow(img)
plt.show()
```

```python
imsave('icon.png', img)
```
